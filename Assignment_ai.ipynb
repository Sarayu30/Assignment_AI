If any, methods like the following can be used to handle missing values across different data types:

### 1. **Numerical Data (int/float)**
   - **Mean/Median/Mode Imputation**: Replace missing values with the column's mean, median, or mode.
   - **Forward/Backward Fill**: Use the previous or next valid observation to fill missing values (`ffill`, `bfill`).
   - **Interpolation**: Estimate missing values by interpolating between existing values.
   - **Replace with a specific value**: Use a constant like 0 or -1 for missing values.

### 2. **Categorical Data (object)**
   - **Mode Imputation**: Replace missing values with the most frequent category (mode).
   - **Replace with 'Unknown'**: Introduce a new category for missing values.
   - **Forward/Backward Fill**: Use nearby values to fill the gaps.

### 3. **Text Data (object)**
   - **Replace with Empty String**: Use an empty string (`''`) to indicate missing text.
   - **Replace with 'Unknown'**: Use a placeholder word like 'Unknown'.

### 4. **Datetime Data**
   - **Replace with a specific date**: Use a default date (e.g., January 1, 1900) for missing values.
   - **Interpolate/Fill**: Forward or backward fill based on nearby timestamps.

### 5. **General Approach** 
   - **Drop Rows/Columns**: Remove rows or columns with missing values if they are not critical.
   - **Model-based Imputation**: Use predictive models (e.g., KNN, regression) to estimate missing values.
   - **Flag and Impute**: Create a binary column to indicate missing values and fill them.

```python
#3.Text Cleaning and Processing
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download NLTK resources
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    # Lowercase
    text = text.lower()
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Tokenize
    tokens = text.split()
    # Remove stopwords and lemmatize
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return ' '.join(tokens)

# Apply cleaning to relevant columns
webtoons_data['clean_details'] = webtoons_data['details'].apply(clean_text)
webtoons_data['clean_genres'] = webtoons_data['genres'].apply(clean_text)
webtoons_data['clean_tags'] = webtoons_data['tags'].apply(clean_text)

# Optionally, combine features into a single text feature
webtoons_data['combined_text'] = webtoons_data['clean_details'] + ' ' + webtoons_data['clean_genres'] + ' ' + webtoons_data['clean_tags']
```

```python
#4.Building the Content Classifier using  using Python and scikit-learn.
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
df = webtoons_data.copy()

# Combine relevant columns, including the 'description', into a single 'features' column
df['features'] = df['title'] + ' ' + df['creators']  + ' ' + df['details'].fillna('')

# Step 1: Vectorize the text using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['features'])

# Step 2: Calculate cosine similarity between webtoons
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Step 3: Create a function to classify webtoons based on similarity
def classify_webtoons(webtoon_index, top_n=15):
    similarity_scores = list(enumerate(cosine_sim[webtoon_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    similar_webtoons = similarity_scores[1:top_n+1]  # Exclude the first one (itself)
    
     
    print(f"Top 15 Anime Like: {df['title'][webtoon_index]}")
    
    for idx, score in similar_webtoons:
        print(f"Title: {df['title'][idx]}, Similarity Score: {score:.2f}")
        
# Step 4: Classify webtoons and print the top 15 similar ones
# Example: Classifying webtoons similar to the first one in the dataset
classify_webtoons(webtoon_index=0, top_n=15)
# Loop through all webtoons in the dataset and get top 15 similar webtoons
for i in range(len(df)):
    print(f"\n\n===== Webtoon {i+1} =====")
    classify_webtoons(webtoon_index=i, top_n=15)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from textblob import TextBlob

# Sample comments generated manually
comments = [
    "I absolutely loved this webtoon! The art is beautiful and the story is so engaging!",
    "This is the best thing I've read in a while. Highly recommend it!",
    "Not my cup of tea. The plot seems dull and the characters are uninteresting.",
    "I found it quite funny and enjoyable. Will read more episodes!",
    "Terrible! I expected better from this author.",
    "It's okay, but I feel like it could use more depth.",
    "Amazing work! Can't wait for the next episode!",
    "I'm disappointed. The pacing is too slow and the humor falls flat.",
    "This webtoon is a masterpiece! I'm a huge fan!",
    "I didn't like it at all. The storyline is confusing."
]

# Function to perform sentiment analysis
def analyze_comments(comments):
    positive_comments = 0
    negative_comments = 0
    neutral_comments = 0
    total_polarity = 0
    
    for comment in comments:
        analysis = TextBlob(comment)
        polarity = analysis.sentiment.polarity
        total_polarity += polarity
        
        # Classify the sentiment based on polarity
        if polarity > 0:
            positive_comments += 1
        elif polarity < 0:
            negative_comments += 1
        else:
            neutral_comments += 1

    total_comments = len(comments)
    positive_percentage = (positive_comments / total_comments) * 100
    negative_percentage = (negative_comments / total_comments) * 100
    neutral_percentage = (neutral_comments / total_comments) * 100
    average_polarity = total_polarity / total_comments

    return positive_percentage, negative_percentage, neutral_percentage, average_polarity

# Perform sentiment analysis
positive_percentage, negative_percentage, neutral_percentage, average_polarity = analyze_comments(comments)

# Summary of the results
print(f"Positive comments: {positive_percentage:.2f}%")
print(f"Negative comments: {negative_percentage:.2f}%")
print(f"Neutral comments: {neutral_percentage:.2f}%")
print(f"Average sentiment polarity: {average_polarity:.2f}")

# Generate Word Clouds for positive and negative comments
positive_words = ' '.join([comment for comment in comments if TextBlob(comment).sentiment.polarity > 0])
negative_words = ' '.join([comment for comment in comments if TextBlob(comment).sentiment.polarity < 0])

# Create word clouds
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
wordcloud_positive = WordCloud(width=400, height=400, background_color='white').generate(positive_words)
plt.imshow(wordcloud_positive, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Positive Comments')

plt.subplot(1, 2, 2)
wordcloud_negative = WordCloud(width=400, height=400, background_color='white').generate(negative_words)
plt.imshow(wordcloud_negative, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Negative Comments')

plt.tight_layout()
plt.show()

# Plotting sentiment distribution
labels = ['Positive', 'Negative', 'Neutral']
sizes = [positive_percentage, negative_percentage, neutral_percentage]
colors = ['#66c2a5', '#fc8d62', '#8da0cb']

plt.figure(figsize=(8, 6))
plt.bar(labels, sizes, color=colors)
plt.ylabel('Percentage (%)')
plt.title('Sentiment Distribution of Comments')
plt.ylim(0, 100)
plt.show()
```

```python
import re
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from collections import Counter

# Ensure necessary NLTK resources are downloaded
nltk.download('punkt')

# Function to clean and summarize content
def summarize_content(content):
    # Remove chapter headings and non-narrative parts
    cleaned_content = re.sub(r'Chapter.*?\n', '', content)  # Remove chapter headings
    cleaned_content = re.sub(r'\n+', ' ', cleaned_content)  # Replace multiple newlines with a single space
    cleaned_content = re.sub(r'\s+', ' ', cleaned_content)  # Replace multiple spaces with a single space

    # Tokenize the cleaned content into sentences
    sentences = sent_tokenize(cleaned_content)

    # Tokenize into words and calculate word frequencies
    words = word_tokenize(cleaned_content.lower())  # Convert to lower case for uniformity
    word_freq = Counter(words)

    # Score each sentence based on the frequency of words it contains
    sentence_scores = {}
    for sentence in sentences:
        for word in word_tokenize(sentence.lower()):
            if word in word_freq:
                if sentence not in sentence_scores:
                    sentence_scores[sentence] = word_freq[word]
                else:
                    sentence_scores[sentence] += word_freq[word]

    # Get the top 5 sentences based on their scores
    summary_sentences = sorted(sentence_scores, key=sentence_scores.get, reverse=True)[:5]

    return ' '.join(summary_sentences)  # Join the selected sentences into a summary

 
# Generate summary
summary = summarize_content(content)
print("Summary:")
print(summary)
```

```python
import re
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk import ne_chunk
from nltk.tree import Tree
from collections import Counter
from itertools import combinations

# Ensure necessary NLTK resources are downloaded
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

# Function to extract all characters and their co-occurrences
def extract_characters_and_co_occurrences(content):
    # Basic cleaning (optional)
    content = re.sub(r'\s+', ' ', content)  # Replace multiple whitespace with single space

    # Tokenize into sentences
    sentences = sent_tokenize(content)

    # POS tagging
    tagged_sentences = [nltk.pos_tag(word_tokenize(sentence)) for sentence in sentences]

    # Function to extract person names
    def extract_person_names(tagged_sentences):
        person_names = []
        for tagged_sentence in tagged_sentences:
            chunks = ne_chunk(tagged_sentence)
            for chunk in chunks:
                if isinstance(chunk, Tree) and chunk.label() == 'PERSON':
                    name = ' '.join([token for token, pos in chunk.leaves()])
                    person_names.append(name)
        return person_names

    # Extract person names
    person_names = extract_person_names(tagged_sentences)

    # Frequency analysis
    name_counts = Counter(person_names)
    all_characters = list(name_counts.keys())  # Get all character names
    main_characters = [name for name, count in name_counts.items() if count > 5]  # Define main characters based on frequency

    # Co-occurrence analysis
    character_pairs = []
    for tagged_sentence in tagged_sentences:
        chunked = ne_chunk(tagged_sentence)
        persons_in_sentence = [
            ' '.join([token for token, pos in chunk.leaves()])
            for chunk in chunked
            if isinstance(chunk, Tree) and chunk.label() == 'PERSON'
        ]
        for pair in combinations(set(persons_in_sentence), 2):
            character_pairs.append(pair)

    pair_counts = Counter(character_pairs)

    # Return results
    return all_characters, main_characters, pair_counts.most_common(5)  # Get the top 5 pairs

 

# Get characters and their co-occurrences
all_characters, main_characters, top_pairs = extract_characters_and_co_occurrences(content)

# Function to generate responses based on keywords
def chatbot_response(user_input):
    # Tokenize user input
    tokens = word_tokenize(user_input.lower())
    
    # Keywords and their corresponding responses
    keywords_responses = {
        "about": f"Castle Swimmer' is a webtoon that involves {summary}...",
        "characters": "They are:\n" + "\n".join(all_characters),
        "main": "They are:\n" + "\n".join(main_characters),
        "co-occurrences": "Top Co-occurring Characters:\n" + "\n".join(
            f"{pair[0]} & {pair[1]}: {count} co-occurrences" for (pair, count) in top_pairs
        ),
        "journey": "The journey is full of unexpected twists and self-discovery.",
        "themes": "The webtoon explores themes of friendship, bravery, and destiny.",
        "details": content  # Return the full content
    }
    
    # Check for keywords in user input
    for keyword, response in keywords_responses.items():
        if keyword in tokens:
            return response
            
    return "I'm sorry, I don't have information on that."

# Chatbot loop
print("Welcome to the Castle Swimmer Chatbot! Ask me anything about the webtoon.")
while True:
    user_input = input("You: ")
    if user_input.lower() in ['exit', 'quit']:
        print("Chatbot: Goodbye!")
        break
    response = chatbot_response(user_input)
    print("Chatbot:", response)

 
```

## Purpose
The **Castle Swimmer Chatbot** is designed to answer user queries about the webtoon *Castle Swimmer*. It utilizes Natural Language Processing (NLP) techniques to extract character information and analyze their relationships.

## Key Components

1. **Imports and Setup**:
   - The chatbot uses the `nltk` library for NLP tasks, including tokenization, part-of-speech tagging, and named entity recognition.
   - Necessary NLTK resources are downloaded for processing.

2. **Content Extraction and Co-occurrence Analysis**:
   - **Function**: `extract_characters_and_co_occurrences(content)`
     - Cleans the input content.
     - Tokenizes the text into sentences.
     - Uses POS tagging to identify characters (persons).
     - Extracts character names and counts their occurrences.
     - Analyzes co-occurrences of character pairs within sentences.
     - Returns:
       - All character names.
       - Main characters (those mentioned frequently).
       - Top co-occurring character pairs.

3. **Chatbot Response Generation**:
